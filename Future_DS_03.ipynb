```python
#making colums in a proper format
data.columns = (
    data.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)
```

```python
data.columns
```

```python
rating_columns = [
    "well_versed_with_the_subject",
    "explains_concepts_in_an_understandable_way",
    "use_of_presentations",
    "degree_of_difficulty_of_assignments",
    "solves_doubts_willingly",
    "structuring_of_the_course",
    "provides_support_for_students_going_above_and_beyond",
    "course_recommendation_based_on_relevance"
]
```

```python
data[rating_columns].isnull().sum()
```

```python
data["total_rating"]=data[rating_columns].mean(axis=1)
```

```python
data
```

```python
data.total_rating.describe()
```

```python
values=data["total_rating"].round().value_counts().sort_index()
```

```python
values
```

```python
#Graphs of TotalRatings
values.plot(kind="bar")
plt.xlabel("Total Rating")
plt.ylabel("Count")
plt.title("Distribution of Total Ratings")
plt.show()
```

```python
value2=data[rating_columns].mean().sort_values(ascending=False)
```

```python
value2
```

```python
#Category wise comparision
value2.plot(kind="bar")
plt.xlabel("Rating Category")
plt.ylabel("Average Rating")
plt.title("Average Ratings for Different Categories")
plt.xticks(rotation=60,ha='right')
plt.show()
```
